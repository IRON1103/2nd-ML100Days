```python
import os
import numpy as np
import pandas as pd
f_app = os.path.join('./data/', 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
```

##### 資料的 row 數以及 column 數

```python
print('len : Column Count = %i, Row Count = %i' % (len(app_train.columns), (len(app_train.index))))
print('shape : Column Count = %i, Row Count = %i' % (app_train.shape[1], app_train.shape[0]))
```

##### 列出所有欄位

```python
print('%s' %(app_train.columns.tolist()))# 只有欄位名稱, 沒有型態
```

##### 截取部分資料

```python
print(app_train[100:102]) # [r1:r2] 從第r1列開始取到第r2-1列
```

```python
filter_1 = (app_train['CODE_GENDER']=='F')# 取得 CODE_GENDER 等於 F 的過濾器
filter_2 = (app_train['FLAG_OWN_CAR']=='Y')# 取得 FLAG_OWN_CAR 等於 Y 的過濾器
print(app_train[filter_1 & filter_2])# 只列出滿足過濾器的資料列
```

```python
print(len(app_train[filter_1].index)) # 滿足過濾器的 row 數
print(len(app_train[filter_2].index)) # 滿足過濾器的 row 數
print(len(app_train[filter_1 & filter_2].index)) # 滿足過濾器的 row 數
```
