## 作業3：依照 Day_001_example_of_metrics.ipynb 完成 Mean Squared Error 的函式

```python
import numpy as np
import matplotlib.pyplot as plt
w = 3
b = 0.5

x_lin = np.linspace(0, 100, 101) # 等差數列
y = (x_lin + np.random.randn(101) * 5) * w + b # 標準正態分佈

plt.plot(x_lin, y, 'b.', label = 'data points') # X軸資料，Y軸資料，藍點，圖例文字
plt.title("Assume we have data points") # 標題
plt.legend(loc = 2) # 圖例位置，2=左上角
plt.show()

y_hat = x_lin * w + b
plt.plot(x_lin, y, 'b.', label = 'data')
plt.plot(x_lin, y_hat, 'r-', label = 'prediction')
plt.title("Assume we have data points (And the prediction)")
plt.legend(loc = 2)
plt.show()
```

$ MAE = \frac{\sum_{i=1}^{n} |{Y_i - X_i}|}{n} $

```python
def mean_absolute_error(y, yp):
    """
    計算 MAE
    Args:
        - y: 實際值
        - yp: 預測值
    Return:
        - mae: MAE
    """
    mae = sum(abs(y - yp)) / len(y)
    return mae

MAE = mean_absolute_error(y, y_hat)
print("The Mean absolute error is %.3f" % (MAE))
```

$ MSE = \frac{1}{n}\sum_{i=1}^{n}{(Y_i - \hat{Y}_i)^2} $
    

```python
def mean_squared_error(y, yp):
    """
    計算 MSE
    Args:
        - y: 實際值
        - yp: 預測值
    Return:
        - mae: MSE
    """
    mse = sum((y-yp)**2) / len(y)
    return mse

MSE = mean_squared_error(y, y_hat)
print("The Mean squared error is %.3f" % (MSE))
```
